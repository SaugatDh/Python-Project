```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
```

```python
print(df.head())
```

```python
sns.scatterplot(x="cgpa",y="resume_score",hue="placed",data=df)
```

```python
X=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron

p=Perceptron()
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values,y.values,clf=p,legend=2)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X,y)

clf.coef_
```

```python
plot_decision_regions(X.values,y.values,clf=clf,legend=2)
```

```python
from sklearn.metrics import accuracy_score

y_pred = p.predict(X)
print("Training Accuracy:", accuracy_score(y, y_pred))
```

```python
from sklearn.metrics import accuracy_score

y_pred = clf.predict(X)
print("Training Accuracy:", accuracy_score(y, y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

p = make_pipeline(StandardScaler(), Perceptron())
p.fit(X,y)
print("Scaled Perceptron Accuracy:", accuracy_score(y, p.predict(X)))
plot_decision_regions(X.values,y.values,clf=p,legend=2)

```

```python
clf = make_pipeline(StandardScaler(), LogisticRegression())

clf.fit(X,y)
print("Scaled Perceptron Accuracy:", accuracy_score(y, clf.predict(X)))
plot_decision_regions(X.values,y.values,clf=clf,legend=2)
```
